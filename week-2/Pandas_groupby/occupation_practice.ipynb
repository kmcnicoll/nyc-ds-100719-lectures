### Step 3. Assign it to a variable called users.

```python
users=pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep="|")
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation").mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users.groupby("occupation")["gender"].value_counts()
```

```python
new = users.groupby(['occupation','gender'])['user_id'].count().reset_index()
new['ratio'] = new.groupby('occupation')['user_id'].apply(lambda x: x/x.sum())
ratio = new[new['gender'] == 'M'].sort_values(by='ratio',ascending=False)[['occupation','ratio']]
ratio.reset_index().drop('index',axis=1)
ratio
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby(["occupation", "gender"]).aggregate({"age":[max, min]})
```

```python
users.groupby(["occupation", "gender"]).aggregate({"spirit_servings":[max, min], "wine_servings":["max","min", "mean"]})
```

```python
users.groupby(["occupation","gender"])["age"].mean()
```
